```python
import os, subprocess, sys

REPO = '/content/z-retina'  # upload the repo here or clone it
assert os.path.isdir(REPO), f'Expected repo at {REPO}'
os.chdir(REPO)
# Colab: install Poetry, then project deps (includes iterative-stratification for research split)
subprocess.check_call('curl -sSL https://install.python-poetry.org | python3 -', shell=True)
poetry = os.path.expanduser('~/.local/bin/poetry')
subprocess.check_call([poetry, 'install', '--no-interaction'], cwd=REPO)
```

```python
CONFIG = os.path.join(REPO, 'configs', 'default.yaml')
CHECKPOINT = os.path.join(REPO, 'runs', 'demo_caformer_b36', 'best.pth')

assert os.path.isfile(CONFIG), f'Missing config: {CONFIG}'
assert os.path.isfile(CHECKPOINT), f'Missing checkpoint: {CHECKPOINT}'
```

```python
import subprocess

poetry_bin = os.path.expanduser('~/.local/bin/poetry')
code = (
    "from z_retina.evaluate import evaluate_table_s1, print_metrics; "
    f"m=evaluate_table_s1({CONFIG!r},{CHECKPOINT!r},use_tta=True); print_metrics(m)"
)
subprocess.check_call([poetry_bin, 'run', 'python', '-c', code], cwd=REPO)
```
